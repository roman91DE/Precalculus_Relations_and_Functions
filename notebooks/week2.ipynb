```python
"""
Question 7
An arrow is shot into the air. Its height is given by the function 

y(x)=64x+46−16x^2

What is the maximum height (in yards) that the arrow reaches?
"""

x = sympy.symbols("x")
y = 64*x + 46 - 16*x**2
y_prime = sympy.diff(y, x)
critical_points = sympy.solve(y_prime, x)
max_height = max(y.subs(x, cp) for cp in critical_points if cp.is_real)
print(f"Maximum height of the arrow: {max_height:.2f} yards")
```

```python
"""
The Throckmorton Company models its monthly revenue (in dollars) from the sale of 
x items with the function 

R(x)=35x−0.25x 

The company has a fixed cost of $540 per month and a variable cost of $3 per item. 
How many items must the company sell to break even?


"""

R = 35 * x - 0.25 * x**2
C = 540 + 3 * x 
break_even_eq = sympy.Eq(R, C)
break_even_solutions = sympy.solve(break_even_eq, x)
break_even_solutions = [s.evalf() for s in break_even_solutions if s.is_real and s >= 0]
print(f"Break-even solutions: {break_even_solutions}")

```

```python
"""
Question 9 (Composition of Functions)
Let f(x) = 3x^2 + 2, g(x) = √(x + 1), and h(x) = 1/x. Find an expression for (f∘g∘h)(x) for x > 0.
"""

f, g, h, x = sympy.symbols('f g h x')
f_expr = 3*x**2 + 2
g_expr = sympy.sqrt(x + 1)
h_expr = 1/x
composed_expr = f_expr.subs(x, g_expr.subs(x, h_expr))
composed_expr = sympy.simplify(composed_expr)
composed_expr
```

```python
"""
Question 10 '(Inverse Functions)
Let f(x) = (12x - 1) / (2x + 4). Find the inverse function f^(-1)(x).
"""
x, y = sympy.symbols('x y')
f = sympy.Eq(
    y,
    ((12*x) - 1) / ((2*x) + 4)
)

inverse = sympy.solve(f, x)[0].subs(y, x)  # Solve for x in terms of y and substitute y with x
inverse
```
