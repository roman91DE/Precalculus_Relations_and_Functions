```python
from sympy import *
import numpy as np
```

```python
x, y = symbols("x,y")
```

```python
expr1 = Eq((x**2) + (y**2), 9)
expr1
```

```python
sympify(solve(expr1, y))  # Two Solutions
```

```python
plot_implicit(
    expr1,
    (x, -4, 4),
    (y, -4, 4),
    title=str(expr1),
    xlabel="x",
    ylabel="y",
    line_color="blue",
)
```

```python
expr2 = Eq(Abs(y), 3 * x)
expr2  # Cant Solve for Y
```

```python
plot_implicit(
    expr2,
    (x, -4, 4),
    (y, -4, 4),
    title=str(expr2),
    xlabel="x",
    ylabel="y",
    line_color="blue",
)
```

```python
expr3 = Eq(x - y, 3)
solve(expr3, y)
```

```python
plot_implicit(
    expr3,
    (x, -5, 5),
    (y, -5, 5),
    title=str(expr3),
    xlabel="x",
    ylabel="y",
    line_color="green",
)
```

```python
g, x, y = symbols("g,x,y")

expr = Eq(Function(g)(x), x**2 + 4 * x - 1)

expr
```

```python
h, x = symbols("h,x")

lhs = Function(h)(x)
rhs = x / (x**2 + 1)

expr = Eq(lhs, rhs)
expr
```

```python
plot(rhs, (x, -10, 10), xlabel=False, ylabel=False)
```

```python
for n in np.linspace(-2.0, 2.0, 20):
    print(f"h({n:>5.2f}) = {rhs.replace(x, n):>5.2f}")
```

## Function Basics


```python
f, x = symbols("f,x")
lhs = Function(f)(x)
rhs = 3 * x**2 - 1
expr = Eq(lhs, rhs)

subExpr = expr.subs({x: x + 2})

Eq(subExpr.lhs, expand(subExpr.rhs, hints="mul"))
```

```python
x, y = symbols("x,y")

eq1 = Eq(x - y, 1)
assert len(solve(eq1, y)) == 1


eq2 = Eq((x - 1) ** 2, 2 - y**2)

"""
(x-1)^2     = 2 - y^2
x^2 -2x +1  = 2 - y^2
x^2 -2x -1  = - y^2
-x^2 +2x +1 = y^2

y   = sqrt(-x^2 +2x +1)
y   = 


solve -x^2 +2x +1:
---
0 = -x^2 +2x +1
0 = x^2 -2x -1
p = -2 ; q = -1 

x1/x2 = -(-2/2) +- sqrt( (-2/2)^2 +1 )
      = 1 +- sqrt(2)
x1 = 1 + sqrt(2)
x2 = 1 - sqrt(2)

"""

solve(eq2, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1 - np.sqrt(2), 1 + np.sqrt(2), 400)
y = np.sqrt(-(x**2) + 2 * x + 1)

plt.plot(x, y, label="+sqrt(...)")
plt.plot(x, -y, label="-sqrt(...)")
plt.title("y = (-x + 2x + 1)")
plt.axhline(0, color="gray", linestyle="--")
plt.axvline(1 - np.sqrt(2), color="red", linestyle=":")
plt.axvline(1 + np.sqrt(2), color="red", linestyle=":")
plt.legend()
plt.grid()
plt.show()
```

```python
"""
x/y  = 5
x    = 5y
y    = x/5
"""

x, y = symbols("x,y")

expr = Eq(Function(y)(x), x / 5)

plot(expr.rhs, (x, -10, 10))
expr
```

```python
x, y = symbols("x,y")

"""
2xy +1 = 7
2xy    = 6
y      = 6 / 2x
"""


expr = Eq(Function(y)(x), 6 / (2 * x))

plot(expr.rhs, (x, -10, 10))
expr
```
