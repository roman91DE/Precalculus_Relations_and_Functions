```python
from functools import partial
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
import seaborn as sns
```

```python
def linear(x: float, m: float, b: float) -> float:
    return m * x + b


# --- Partial functions ---
f1 = partial(linear, m=2, b=1)
f2 = partial(linear, m=4, b=-2)
f3 = partial(linear, m=-1, b=10)
f4 = partial(linear, m=0.5, b=0)

df = pd.DataFrame({"x": np.arange(-100, 101, 1, dtype=np.int32)})
df["f(x) = 2x+1"] = df["x"].map(f1)
df["f(x) = 4x-2"] = df["x"].map(f2)
df["f(x) = -x+10"] = df["x"].map(f3)
df["f(x) = 0.5x"] = df["x"].map(f4)

# --- Plot without melting ---
fig, axes = plt.subplots(2, 2, figsize=(12, 8), sharex=True, sharey=True)

columns = [
    ("f(x) = 2x+1", axes[0, 0]),
    ("f(x) = 4x-2", axes[0, 1]),
    ("f(x) = -x+10", axes[1, 0]),
    ("f(x) = 0.5x", axes[1, 1]),
]

for col_name, ax in columns:
    sns.pointplot(x=df["x"], y=df[col_name], ax=ax, native_scale=True)
    ax.set_title(col_name)
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
# --- Define quadratic function ---
def quadratic(x: float, a: float, b: float, c: float) -> float:
    return a * x**2 + b * x + c


# --- Partial functions ---
f1 = partial(quadratic, a=1, b=0, c=0)  # f(x) = x
f2 = partial(quadratic, a=-1, b=0, c=0)  # f(x) = -x
f3 = partial(quadratic, a=4, b=1, c=0)  # f(x) = x + 2x + 1
f4 = partial(quadratic, a=0.25, b=-1, c=0)  # f(x) = 0.5x - 3x + 2

# --- Create dataframe ---
df = pd.DataFrame({"x": np.arange(-100, 101, 1, dtype=np.int32)})

# --- Apply functions ---
df["f(x) = x"] = df["x"].map(f1)
df["f(x) = -x"] = df["x"].map(f2)
df["f(x) = 4x+x"] = df["x"].map(f3)
df["f(x) = 0.25x+x"] = df["x"].map(f4)

# --- Plot without melting ---
fig, axes = plt.subplots(2, 2, figsize=(12, 8), sharex=True, sharey=True)

columns = [
    ("f(x) = x", axes[0, 0]),
    ("f(x) = -x", axes[0, 1]),
    ("f(x) = 4x+x", axes[1, 0]),
    ("f(x) = 0.25x+x", axes[1, 1]),
]

for col_name, ax in columns:
    sns.pointplot(x=df["x"], y=df[col_name], ax=ax, native_scale=True)
    ax.set_title(col_name)
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
# --- Define cubic function ---
def cubic(x: float, a: float, b: float, c: float, d: float) -> float:
    return a * x**3 + b * x**2 + c * x + d


# --- Partial functions ---
f1 = partial(cubic, a=1, b=0, c=0, d=0)  # f(x) = x
f2 = partial(cubic, a=-1, b=0, c=0, d=0)  # f(x) = -x
f3 = partial(cubic, a=1, b=2, c=0, d=0)  # f(x) = x + 2x
f4 = partial(cubic, a=0.25, b=-1, c=0, d=5)  # f(x) = 0.25x - x + 5

# --- Create dataframe ---
df = pd.DataFrame({"x": np.arange(-100, 101, 1, dtype=np.int32)})

# --- Apply functions ---
df["f(x) = x"] = df["x"].map(f1)
df["f(x) = -x"] = df["x"].map(f2)
df["f(x) = x+2x"] = df["x"].map(f3)
df["f(x) = 0.25x-x+5"] = df["x"].map(f4)

# --- Plot without melting ---
fig, axes = plt.subplots(2, 2, figsize=(12, 8), sharex=True, sharey=True)

columns = [
    ("f(x) = x", axes[0, 0]),
    ("f(x) = -x", axes[0, 1]),
    ("f(x) = x+2x", axes[1, 0]),
    ("f(x) = 0.25x-x+5", axes[1, 1]),
]

for col_name, ax in columns:
    sns.pointplot(x=df["x"], y=df[col_name], ax=ax, native_scale=True)
    ax.set_title(col_name)
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
# --- Define square root function ---
def square_root(x: float, a: float, b: float) -> float:
    return a * np.sqrt(x) + b


# --- Partial functions ---
f1 = partial(square_root, a=1, b=0)  # f(x) = sqrt(x)
f2 = partial(square_root, a=2, b=1)  # f(x) = 2*sqrt(x) + 1
f3 = partial(square_root, a=-1, b=5)  # f(x) = -sqrt(x) + 5
f4 = partial(square_root, a=0.5, b=-2)  # f(x) = 0.5*sqrt(x) - 2

# --- Create dataframe ---
df = pd.DataFrame(
    {"x": np.arange(0, 101, 1, dtype=np.int32)}
)  # Only non-negative x for sqrt

# --- Apply functions ---
df["f(x) = sqrt(x)"] = df["x"].map(f1)
df["f(x) = 2sqrt(x)+1"] = df["x"].map(f2)
df["f(x) = -sqrt(x)+5"] = df["x"].map(f3)
df["f(x) = 0.5sqrt(x)-2"] = df["x"].map(f4)

# --- Plot without melting ---
fig, axes = plt.subplots(2, 2, figsize=(12, 8), sharex=True, sharey=True)

columns = [
    ("f(x) = sqrt(x)", axes[0, 0]),
    ("f(x) = 2sqrt(x)+1", axes[0, 1]),
    ("f(x) = -sqrt(x)+5", axes[1, 0]),
    ("f(x) = 0.5sqrt(x)-2", axes[1, 1]),
]

for col_name, ax in columns:
    sns.pointplot(x=df["x"], y=df[col_name], ax=ax, native_scale=True)
    ax.set_title(col_name)
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
# --- Define fractional function ---
def fractional(x: float, a: float, b: float) -> float:
    return a / x + b


# --- Partial functions ---
f1 = partial(fractional, a=1, b=0)  # f(x) = 1/x
f2 = partial(fractional, a=-1, b=2)  # f(x) = -1/x + 2
f3 = partial(fractional, a=5, b=-3)  # f(x) = 5/x - 3
f4 = partial(fractional, a=0.5, b=1)  # f(x) = 0.5/x + 1

# --- Create dataframe ---
# Avoid x = 0 to prevent division by zero!
df = pd.DataFrame({"x": np.linspace(-10.0, 11.0, 1000, dtype=np.float32)})
df = df[df["x"] != 0]  # Remove zero manually

# --- Apply functions ---
df["f(x) = 1/x"] = df["x"].map(f1)
df["f(x) = -1/x+2"] = df["x"].map(f2)
df["f(x) = 5/x-3"] = df["x"].map(f3)
df["f(x) = 0.5/x+1"] = df["x"].map(f4)

# --- Plot without melting ---
fig, axes = plt.subplots(2, 2, figsize=(12, 8), sharex=True, sharey=True)

columns = [
    ("f(x) = 1/x", axes[0, 0]),
    ("f(x) = -1/x+2", axes[0, 1]),
    ("f(x) = 5/x-3", axes[1, 0]),
    ("f(x) = 0.5/x+1", axes[1, 1]),
]

for col_name, ax in columns:
    sns.pointplot(x=df["x"], y=df[col_name], ax=ax, native_scale=True)
    ax.set_title(col_name)
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
"""
f(x) = cos(x)

"""


# Define x and f (f is a Function, not evaluated yet)
x = symbols("x")
f = Function("f")

# Now use f(x) inside the equation
expr = Eq(f(x), cos(x))

# Differentiate both sides
diff_expr = Eq(diff(expr.lhs, x), diff(expr.rhs, x))

zexpr = Eq(diff_expr.lhs + diff_expr.rhs, 0)

dsolve(zexpr)
```
