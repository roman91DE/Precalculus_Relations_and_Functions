```python
import numpy as np
import sympy as sp
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
f, g, h, x, y = sp.symbols('f g h x y')
```

```python
# q1
expr = (1/3) ** x
sp.plot(
    expr,
    (x, -2, 2),
    title='Plot of (1/3)^x',
    xlabel='x',
    ylabel='y'
)

plt.show()
```

```python
# q2
num = 2 ** (-x)
denom = ((2 ** x) - 2) * ((2 ** x) - 4)
f_expr = num / denom
sp.plot(
    f_expr,
    (x, -4, 4),
    xlabel='x',
    ylabel='f(x)'
)

plt.show()
```

```python
# q3
g_expr = 3 - (5 ** (-x))
sp.plot(
    g_expr,
    (x, -4, 4),
    xlabel='x',
    ylabel='g(x)'
)
plt.show()

for v in range(-10, 20, 2):
    print(f"g({v:>3}) = {float(g_expr.subs(x, v))}")
```
