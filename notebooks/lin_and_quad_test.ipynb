```python
import math

from sympy import *
import numpy as np
from matplotlib import pyplot as plt
```

```python
X = np.arange(-8, 8)
f1 = lambda x: 5 * x + 1
f2 = lambda x: 5 * x + 8

Y1, Y2 = f1(X), f2(X)

plt.plot(X, Y1)
plt.plot(X, Y2)
plt.plot(-1, 3, "ro", label="Point (-1, 3)")  # 'ro' = red circle
plt.grid(True)
plt.legend()
plt.axis("equal")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
f3 = lambda x: 2 * x - 3
f4 = lambda x: (-1 / 2) * x + 5
f4 = lambda x: (-1 / 2) * (x - 2) + 4

Y3, Y4 = f3(X), f4(X)

plt.plot(X, Y3)
plt.plot(X, Y4)
plt.plot(2, 4, "ro", label="Point (2, 4)")  # 'ro' = red circle
plt.grid(True)
plt.legend()
plt.axis("equal")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
f5 = lambda x: (1 / 8) * x + 42 / 8

assert f5(6) == 6
assert f5(-2) == 5
```

```python
f6 = lambda x: 80 * x - 240
X2 = np.arange(0, 5)
Y5 = f6(X2)

plt.plot(X2, Y5)
plt.plot(3, 0, "ro", label="Surface")  # 'ro' = red circle
plt.grid(True)
plt.legend()
plt.xlabel("Time in Mins")
plt.ylabel("Height")
plt.show()
```
